```python
import pandas as pd
import numpy as np
```

```python
junctions = pd.DataFrame(np.random.randint(50,100,size=(1000, 4)), columns=['a','b','c','d'])
```

```python
junctions_lanes_a = pd.DataFrame(np.random.randint(5,15,size=(1000, 8)), columns=['a_lane_input_1','a_lane_output_1','a_lane_input_2','a_lane_output_2','a_lane_input_3','a_lane_output_3','a_lane_input_4','a_lane_output_4'])
```

```python
junctions_lanes_b = pd.DataFrame(np.random.randint(5,15,size=(1000, 8)), columns=['b_lane_input_1','b_lane_output_1','b_lane_input_2','b_lane_output_2','b_lane_input_3','b_lane_output_3','b_lane_input_4','b_lane_output_4'])
```

```python
junctions_lanes_c = pd.DataFrame(np.random.randint(5,15,size=(1000, 8)), columns=['c_lane_input_1','c_lane_output_1','c_lane_input_2','c_lane_output_2','c_lane_input_3','c_lane_output_3','c_lane_input_4','c_lane_output_4'])
```

```python
junctions_lanes_d = pd.DataFrame(np.random.randint(5,15,size=(1000, 10)), columns=['d_lane_input_1','d_lane_output_1','d_lane_input_2','d_lane_output_2','d_lane_input_3','d_lane_output_3','d_lane_input_4','d_lane_output_4','d_lane_input_5','d_lane_output_5'])
```

```python
result = pd.concat([junctions, junctions_lanes_a,junctions_lanes_b,junctions_lanes_c,junctions_lanes_d], axis=1)
```

```python
result.to_csv('traffic_dataset.csv') 
```
